```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading and Manipulating Velocity and Reference data

path = './Ghia Results/'

U_ref_100 = pd.read_csv(path+'ghia-100u.dat', delimiter='\t', names=['Values', 'Axis'])
U_ref_400 = pd.read_csv(path+'ghia-400u.dat', delimiter='\t', names=['Values', 'Axis'])
U_ref_1000 = pd.read_csv(path+'ghia-1000u.dat', delimiter='\t', names=['Values', 'Axis'])


V_ref_100 = pd.read_csv(path+'ghia-100v.dat', delimiter='\t', names=['Drop'])
V_ref_100['Axis'] = V_ref_100['Drop'].apply(lambda x: float(x.split()[1]))
V_ref_100['Values'] = V_ref_100['Drop'].apply(lambda x: float(x.split()[0]))

V_ref_400 = pd.read_csv(path+'ghia-400v.dat', delimiter='\t', names=['Drop'])
V_ref_400['Axis'] = V_ref_400['Drop'].apply(lambda x: float(x.split()[1]))
V_ref_400['Values'] = V_ref_400['Drop'].apply(lambda x: float(x.split()[0]))

V_ref_1000 = pd.read_csv(path+'ghia-1000v.dat', delimiter='\t', names=['Drop'])
V_ref_1000['Axis'] = V_ref_1000['Drop'].apply(lambda x: float(x.split()[1]))
V_ref_1000['Values'] = V_ref_1000['Drop'].apply(lambda x: float(x.split()[0]))

U_100 = pd.read_csv('U_Re-100_vertical_mid.csv')[::-1].T.values[0]
U_400 = pd.read_csv('U_Re-400_vertical_mid.csv')[::-1].T.values[0]
U_1000 = pd.read_csv('U_Re-1000_vertical_mid.csv')[::-1].T.values[0]

V_100 = pd.read_csv('V_Re-100_horizontal_mid.csv').T.values[0]
V_400 = pd.read_csv('V_Re-400_horizontal_mid.csv').T.values[0]
V_1000 = pd.read_csv('V_Re-1000_horizontal_mid.csv').T.values[0]


N1 = len(U_100)
N2 = len(U_400)
N3 = len(U_1000)
M1 = len(V_100)
M2 = len(V_400)
M3 = len(V_1000)
```

```python
# U velocity profile vs Vertical Centre Line

plt.figure(figsize=(8, 8), dpi=400)


font1 = {'family':'Times New Roman','color':'Black','size':16}
font2 = {'family':'Times New Roman','color':'Black','size':14}


plt.plot(U_100, np.linspace(0,1,N1), label='Present (Re = 100)', color='blue')
plt.scatter(U_ref_100.Values, U_ref_100.Axis, label='Ghia et. al. (Re = 100)', color='blue', marker="^")

plt.plot(U_400, np.linspace(0,1,N2), label='Present (Re = 400)', color='green')
plt.scatter(U_ref_400.Values, U_ref_400.Axis, label='Ghia et. al. (Re = 400)', color='green', marker="s")

plt.plot(U_1000, np.linspace(0,1,N3), label='Present (Re = 1000)', color='red')
plt.scatter(U_ref_1000.Values, U_ref_1000.Axis, label='Ghia et. al. (Re = 10000)', color='red', marker="*")


plt.ylim(0,1)

plt.legend()

plt.title("U velocity profile vs Vertical centre line\n", fontdict=font1)

plt.xlabel("U Velocity", fontdict=font2)
plt.ylabel("Y", fontdict=font2)

# plt.savefig("U_velocity_profile.jpg")

plt.show()
```

```python
# U velocity profile vs Vertical Centre Line

plt.figure(figsize=(8, 8), dpi=400)

font1 = {'family':'Times New Roman','color':'Black','size':16}
font2 = {'family':'Times New Roman','color':'Black','size':14}


plt.plot(np.linspace(0,1,M1), V_100, label='Present (Re = 100)', color='blue')
plt.scatter(V_ref_100.Axis, V_ref_100.Values, label='Ghia et. al. (Re = 100)', color='blue', marker="^")

plt.plot(np.linspace(0,1,M2), V_400, label='Present (Re = 400)', color='green')
plt.scatter(V_ref_400.Axis, V_ref_400.Values, label='Ghia et. al. (Re = 400)', color='green', marker="s")

plt.plot(np.linspace(0,1,M3), V_1000, label='Present (Re = 1000)', color='red')
plt.scatter(V_ref_1000.Axis, V_ref_1000.Values, label='Ghia et. al. (Re = 1000)', color='red', marker="*")


plt.xlim(0,1)

plt.legend()

plt.title("V velocity profile vs Horizontal centre line\n", fontdict=font1)

plt.xlabel("X", fontdict=font2)
plt.ylabel("V Velocity", fontdict=font2)

# plt.savefig("V_velocity_profile.jpg")

plt.show()
```
